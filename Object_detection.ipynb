```python
import torch
import gradio as gr
from PIL import Image

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')

# Object Detection Function
def detect_objects_image(img):
    results = model(img)
    results.render()  # Draw detections on the image
    return Image.fromarray(results.ims[0])  # Convert to PIL image

# ğŸ¨ Custom CSS for Stylish UI with Centered Elements
custom_css = """
    body {background-color: #121212; color: white; font-family: Arial, sans-serif; text-align: center;}
    .gradio-container {border-radius: 15px; padding: 20px; background: #1e1e1e; text-align: center;}
    .gr-button {background-color: #ff5722 !important; color: white !important; border-radius: 8px; font-size: 16px;}
    .gr-button:hover {background-color: #ff784e !important;}
    h1 {color: #4A90E2 !important; text-align: center;}  /* âœ… Title in Blue */
    h2, h3, p {color: #b0b0b0 !important; text-align: center;}  /* âœ… Subtitle in Grey */
    input[type="file"] {color: white;}

    /* âœ… Light Background for Input & Output Boxes */
    .input-box, .output-box {
        background-color: #F5E6CC !important;  /* Light Beige */
        border-radius: 10px;
        padding: 10px;
        text-align: center;
    }
"""

# ğŸ¯ Gradio Interface with Centered Elements
interface = gr.Interface(
    fn=detect_objects_image,
    inputs=gr.Image(type="pil", label="ğŸ“¥ Upload an Image", elem_classes="input-box"),
    outputs=gr.Image(type="pil", label="ğŸ“¤ Detected Objects", elem_classes="output-box"),
    title="ğŸŒŸ YOLOv5 Object Detection ğŸ¯",
    description="âœ¨ Upload an image and let YOLOv5 detect objects in real-time! ğŸš€",
    theme="compact",
    css=custom_css
)

# Launch Gradio app with a Shareable Link
interface.launch(debug=True, share=True)
```
